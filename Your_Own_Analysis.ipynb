## Gene ontology term-enrichment analysis for CLIP and ribosome profiling

```python
import gzip
```

```python
goa_file = '/content/drive/MyDrive/binfo1-work/goa_mouse.gaf.gz'
mapping_file = '/content/drive/MyDrive/binfo1-work/MOUSE_10090_idmapping.dat.gz'
clip_rden_file = '/content/drive/MyDrive/binfo1-work/clip_rden_positive.csv'
mouse_go_mapping_file = '/content/drive/MyDrive/binfo1-work/mouse_go_mapping.csv'
```

```python
import pandas as pd

# 파일 불러오기
df_clip_rden = pd.read_csv(clip_rden_file, index_col=0)
df_go_map = pd.read_csv(mouse_go_mapping_file)

# 인덱스 문자열 변환 및 버전 제거
df_clip_rden.index = df_clip_rden.index.astype(str).str.replace(r"\.\d+$", "", regex=True)

# 확인
df_clip_rden.head(), df_go_map.head()
```

```python
import numpy as np

# GO term에 해당하는 유전자만 병합
df_merged = pd.merge(df_go_map, df_clip_rden, left_on="ensembl_gene_id", right_index=True)

# GO term별 요약 정보 계산
summary = []

for go_term, group in df_merged.groupby("go_id"):
    gene_ids = group["ensembl_gene_id"].unique()
    if len(gene_ids) < 5:
        continue  # 최소 5개 유전자 포함된 term만

    summary.append({
        "GO_term": go_term,
        "n_genes": len(gene_ids),
        "CLIP_log2FC": np.log2(group["clip_enrichment"].mean()),
        "Ribo_log2FC": np.log2(group["rden_change"].mean())
    })

df_summary = pd.DataFrame(summary)
df_summary["size"] = df_summary["n_genes"] / 2  # 시각화용 점 크기

df_summary.head()
```

```python
from scipy.stats import mannwhitneyu
from statsmodels.stats.multitest import multipletests

clip_pvals = []
rden_pvals = []

for go_term in df_summary["GO_term"]:
    gene_ids = df_merged[df_merged["go_id"] == go_term]["ensembl_gene_id"]

    # 그룹 내 값
    clip_in = df_clip_rden.loc[gene_ids, "clip_enrichment"]
    rden_in = df_clip_rden.loc[gene_ids, "rden_change"]

    # 그룹 외 값
    rest_ids = df_clip_rden.index.difference(gene_ids)
    clip_rest = df_clip_rden.loc[rest_ids, "clip_enrichment"]
    rden_rest = df_clip_rden.loc[rest_ids, "rden_change"]

    # 테스트 수행
    clip_pval = mannwhitneyu(clip_in, clip_rest, alternative='two-sided').pvalue
    rden_pval = mannwhitneyu(rden_in, rden_rest, alternative='two-sided').pvalue

    clip_pvals.append(clip_pval)
    rden_pvals.append(rden_pval)

df_summary["CLIP_pval"] = clip_pvals
df_summary["Ribo_pval"] = rden_pvals

# FDR (max of two pvals)
df_summary["FDR"] = df_summary[["CLIP_pval", "Ribo_pval"]].max(axis=1)
df_summary["log10FDR"] = -np.log10(df_summary["FDR"])
```

```python
df_summary.to_csv("go_enrichment_summary.csv", index=False)
```

```python
import requests
import time

# GO term 리스트 추출
go_ids = df_summary["GO_term"].unique().tolist()

# QuickGO API에서 GO term 이름 매핑
def get_go_term_names(go_ids):
    base_url = "https://www.ebi.ac.uk/QuickGO/services/ontology/go/terms/"
    headers = {"Accept": "application/json"}
    mapping = {}

    # 100개 단위로 나눠서 요청 (API 제한 대응)
    for i in range(0, len(go_ids), 100):
        batch = go_ids[i:i+100]
        joined_ids = ",".join(batch)
        url = base_url + joined_ids
        response = requests.get(url, headers=headers)

        if response.status_code == 200:
            results = response.json()["results"]
            for item in results:
                mapping[item["id"]] = item["name"]
        else:
            print(f"Failed to retrieve: {response.status_code}")
        time.sleep(0.3)  # 요청 간 간격

    return mapping

# 이름 매핑
go_term_names = get_go_term_names(go_ids)

# 이름 열 추가
df_summary["GO_label"] = df_summary["GO_term"].map(go_term_names)

# 확인
df_summary[["GO_term", "GO_label"]].drop_duplicates().head(10)
```

Visualization

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

# 색상 설정
cmap = mpl.colormaps["YlOrRd_r"]
norm = mpl.colors.LogNorm(vmin=1e-20, vmax=1)

# 필터링 강화 (FDR < 0.05) + 점 크기 확대
df_plot = df_summary[(df_summary["FDR"] < 0.05) & (df_summary["n_genes"] >= 10)].copy()
df_plot["size"] = df_plot["n_genes"] * 1.5  # 점 크기 키우기

# 시각화
fig, ax = plt.subplots(figsize=(13, 9))
sc = ax.scatter(
    df_plot["CLIP_log2FC"],
    df_plot["Ribo_log2FC"],
    s=df_plot["size"],
    c=df_plot["FDR"],
    cmap=cmap,
    norm=norm,
    edgecolor='black',
    linewidths=0.5,
    alpha=0.9
)

# 기준선
ax.axhline(0, color="gray", linestyle="--", lw=1)
ax.axvline(0, color="gray", linestyle="--", lw=1)

# 축 및 제목
ax.set_xlabel("Enrichment level of LIN28A-bound CLIP tags (log₂)", fontsize=12)
ax.set_ylabel("Ribosome density change upon Lin28a knockdown (log₂)", fontsize=12)
ax.set_title("A  Gene ontology term-enrichment analysis for CLIP and ribosome profiling", fontsize=14, loc='left')

# 색상 컬러바
cbar = fig.colorbar(sc)
cbar.set_label("Term-specific enrichment confidence (false discovery rate)", fontsize=11)
cbar.set_ticks([1e-20, 1e-16, 1e-12, 1e-8, 1e-4, 1])
cbar.ax.set_yticklabels([f"10$^{{{int(np.log10(t))}}}$" for t in cbar.get_ticks()])

# 크기 범례
for s in [10, 20, 40, 80]:
    ax.scatter([], [], s=s*1.5, label=f"{s} genes", color="gray", alpha=0.4)
ax.legend(title="Gene count", loc="lower left", frameon=True)

# 선택된 GO term만 라벨링 (예시 13개)
highlight_labels = [
    "nucleus", "cytoplasm", "nucleosome assembly", "golgi apparatus", "calcium ion binding",
    "nucleosome", "cell surface", "endoplasmic reticulum lumen", "glycosaminoglycan metabolic process",
    "endoplasmic reticulum membrane", "extracellular region", "endoplasmic reticulum", "mitochondrion"
]

# GO_label이 존재할 경우 라벨링
if "GO_label" in df_plot.columns:
    for _, row in df_plot.iterrows():
        if row["GO_label"].lower() in [x.lower() for x in highlight_labels]:
            label = f"{row['GO_label']} ({int(row['n_genes'])})\nC={row['CLIP_pval']:.1e} R={row['Ribo_pval']:.1e}"
            ax.annotate(label,
                        (row["CLIP_log2FC"], row["Ribo_log2FC"]),
                        textcoords="offset points",
                        xytext=(10, 10),
                        fontsize=9,
                        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.5),
                        arrowprops=dict(arrowstyle="->", lw=0.5, color="black"))

plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
